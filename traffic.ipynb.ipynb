```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Indian_Traffic_Violations.csv")

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Weather_Condition"
)

plt.xlabel("Weather Condition")
plt.ylabel("Number of Violations")
plt.title("Impact of Weather Condition on Traffic Violations")

plt.xticks(rotation=45)  # rotate labels for readability
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Indian_Traffic_Violations.csv")

# Create scatter plot
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Speed_Limit",
    y="Recorded_Speed"
)

# Labels and title
plt.xlabel("Speed Limit")
plt.ylabel("Recorded Speed")
plt.title("Speed Limit vs Recorded Speed")

plt.show()
```

```python
import pandas as pd

df = pd.read_csv("Indian_Traffic_Violations.csv")

total_records = len(df)

over_speed = df[df["Recorded_Speed"] > df["Speed_Limit"]]

over_speed_count = len(over_speed)

percentage = (over_speed_count / total_records) * 100

# Print results
print("Total Records:", total_records)
print("Over-speeding Records:", over_speed_count)
print("Percentage of Over-speeding:", round(percentage, 2), "%")
```

```python
plt.hist(df["Driver_Age"], bins=10)

plt.xlabel("Driver Age")
plt.ylabel("Number of Violations")
plt.title("Driver Age Distribution")

plt.show()
```

```python

violation_counts = df["Violation_Type"].value_counts().head(5)


plt.figure(figsize=(7, 7))
plt.pie(
    violation_counts,
    labels=violation_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Top 5 Traffic Violation Types")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Indian_Traffic_Violations.csv")

# Create a new column: Speed Difference
df["Speed_Difference"] = df["Recorded_Speed"] - df["Speed_Limit"]

# Plot speed difference by weather condition
plt.figure(figsize=(9,5))
sns.boxplot(
    data=df,
    x="Weather_Condition",
    y="Speed_Difference"
)

plt.xlabel("Weather Condition")
plt.ylabel("Speed Difference (Recorded - Limit)")
plt.title("Speed Difference Distribution by Weather Condition")

plt.xticks(rotation=45)
plt.show()
```

```python
# Create new feature: Exceeded Speed (only positive values)
df["Exceeded_Speed"] = (df["Recorded_Speed"] - df["Speed_Limit"]).clip(lower=0)

# Display first few rows
print(df[["Speed_Limit", "Recorded_Speed", "Exceeded_Speed"]].head())
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Indian_Traffic_Violations.csv")

# Set figure size
plt.figure(figsize=(9,5))

# Strip plot (shows individual points)
sns.stripplot(
    data=df,
    x="Payment_Method",
    y="Fine_Amount",
    jitter=True,
    alpha=0.4
)

# Swarm plot (shows density clearly)
sns.swarmplot(
    data=df,
    x="Payment_Method",
    y="Fine_Amount",
    size=3
)

# Labels and title
plt.xlabel("Payment Method")
plt.ylabel("Fine Amount")
plt.title("Fine Amount vs Payment Method (Strip + Swarm Plot)")

plt.show()
```
